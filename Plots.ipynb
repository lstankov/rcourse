Top Tips
------------

1. Know the comands for plot, abline, line, hist, density, and curve to quickly generate graphs quickly for exploration
2. pairs and corrplat will let you look at lots of pairwise relationships quickly
3. Start simply with default plot settings and slowly build up
4. For likert data use jitter to move points around a bit to better see relationships
5. For large data use alpha blending to make points semi transperent to better see where the density of points is (see alpha or rgb functions)
6. For big data use heat maps, [ggobi](http://www.ggobi.org/), [bigvis](https://github.com/hadley/bigvis/blob/master/README.md), or hexbins.
7. Learn ggplots as this seems to be where everyone is heading
8. Many many functions have default plots so try throwing output from a function at plot(). For models this will often be plots of assumptions.
9. Follow the general rule to avoid chartjunk (Tufte, 1983 - Above all else show the data)
10. Know what devices are best to export to (you might watn to export as an svg and do further work in illustrator or inkscape)

```python
%%R
library(ggplot2)
library(lattice)
library(corrgram)
library(vcd)
library(wesanderson)
library(foreign)
library(car)
setwd("/Users/phparker/Dropbox/Rcourse/data/")
pisa <- read.spss("PISA2003.sav", to.data.frame = TRUE, use.value.labels = FALSE)
names(pisa)
#Scatter plot matrix
pairs(pisa[,27:32], pch = ".", cex = 3)
#Correlation plot
corrgram(pisa[,27:32], lower.panel = panel.ellipse, upper.panel = panel.pie, order=TRUE, 
		 main = "Correlation plot\nPisa Psychological variables")
```

Scatter plots, box plots, and Histograms
----------------------

Here I give examples of moving from default graphs to preitter graphs.

```python
%%R
#Scatter Plot - Ugly
with(pisa, plot(SCMAT, ANXMAT))

# Scatter Plot - Pretty
with(pisa, plot(jitter(SCMAT,amount = 2), jitter(ANXMAT,amount = 2), pch = ".", cex = 3, col = "#00000075",
				main = "Scatter Plot of\nSelf-concept and Anxiety", xlab = "Self-concept", ylab = "Anxiety",
				bty="n"))

#Lets add a regression line
abline(lm(pisa$ANXMAT ~ pisa$SCMAT), col = rgb(1,0,0,.6), cex = 2)

#Histogram - Ugly
hist(pisa$ESCS)
#Histogram - Pretty
hist(pisa$ESCS,freq = FALSE, col = "grey", border = NULL)
lines(density(pisa$ESCS, na.rm=TRUE), col = "red")

#Bar plot - ugly
boxplot(SCMAT ~ IMMIG, data = pisa, main = "Boxplot of self-concept by \nImmigrant Status")
#Bar plot - pretty
boxplot(SCMAT ~ IMMIG, data = pisa, main = "Boxplot of self-concept by \nImmigrant Status",
		notch = TRUE, col = c("grey25", "grey50", "grey75") )
```

Mapping Qualitative Data
----------------------------
I don't really have a lot of experience with this but when I have done stuff I have used either mosaic plots or foufold plots

```python
%%R
#Categorical data plots - Mosaic
mosaic(IMMIG ~ yearInSchool | gender, data = pisa, highlighting_fill = 
	   	c("#3FAA7F", "#77A8BA", "#C47D9D", "#A1E0B9", "#F1F799"))

# Fourfold plot on gender by self-report social class
    #Fourfold plots are for two by two contingency tables so we need to make a new variable
pisa$socialClass <- recode(pisa$SSECATEG, "c(1,2) = 1; c(3,4) = 0")
#Next we create a contingency table
myTable <- with(pisa, table(gender, socialClass))
#Then plot it
fourfoldplot(myTable, color = c("#99CCFF", "#6699CC"),
			 conf.level = 0.95,
			 std = c("margins", "ind.max", "all.max"),
			 margin = c(1, 2), space = 0.2, main = NULL,
			 mfrow = NULL, mfcol = NULL)
```

GGPLOTS
----------

SPSS has for sometime tried to use a sort of grammar of graphs logic to their plots. However, SPSS does not really have this sort of flexibility. One of the main people in this framework, Hadley Wickham, however has made a grammar of graphs package for R called ggplots 2. I rarely use it as I am considerably quicker with base graphics BUT ggplots are powerful and beautiful. Below I create a pretty scatter plot and full information bar chart.

The idea behind ggplots is to build graphs in discreet stages or components seperated by '+'. ggplots has a powerful faceting system in which multiple graphs are presented in rows and columns to provide a rich but readable picture of your data. 

```python
%%R
#Scatter plot
library(ggplot2)
print(ggplot(pisa, aes(SCMAT, ANXMAT, color = as.factor(yearInSchool)) ) + 
	geom_point(size = 3, position=position_jitter(width=1,height=.5), alpha = .6) + 
	scale_color_manual(values = wes_palette("GrandBudapest")) +
	scale_x_continuous(name = "Anxiety") + scale_y_continuous(name = "Self-concept") + 
	ggtitle("Self-concept by Anxiety") + 
	theme(plot.title=element_text(family="Times", face="bold", size=20)) +
	guides(color=guide_legend(title="Year in school")) )
# Fancy bar plot
print(ggplot(pisa, aes(as.factor(yearInSchool),SCMAT) )+
	geom_point(position = position_jitter(width=.2,height=.5), alpha = 0.6) +
	scale_x_discrete(breaks= 8:11, labels=c("Yr8", "Yr9", "Yr10", "Yr11"), name = "Year in school") +
	scale_y_continuous(name="Self-concept") +
	ggtitle("Self-concept by year in school") + 
	theme(plot.title=element_text(family="Times", face="bold", size=20)) )
```
